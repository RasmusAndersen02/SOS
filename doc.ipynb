```python
import numpy as np
import pandas as pd
import os
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import lemmy

lemmatizer = lemmy.load("da")
dir = "./files"
txt_files = [os.path.join(dir, f) for f in os.listdir(dir) if f.endswith('.txt')]

keywords = ['kost', 'vitamin', 'kosttilskud', 'bivirkning', 'gravid', 'kostvane']
cols = ['file'] + keywords

df = pd.DataFrame(columns=cols)

def CountKeywords(txt_file):
    with open(txt_file, 'r', encoding='utf-8', errors='ignore') as file:
        s_file = file.read().replace('-', ' ')  

    word_list = word_tokenize(s_file.lower())  

    lemmas = [lemmatizer.lemmatize("", word) for word in word_list]
    lemmas = [lemma[0] for lemma in lemmas]
    print(lemmas)

    df_list_append = [txt_file] + [0] * len(keywords)


    for em in lemmas:
        if em in keywords:
            df_list_append[keywords.index(em) + 1] += 1 

    return df_list_append

for em in txt_files:
    keywords_4file = CountKeywords(em)
    df.loc[len(df)] = keywords_4file


```
